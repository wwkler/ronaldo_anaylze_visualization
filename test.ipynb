```python
# 'Minute'에 35, 57 이렇게 나타나면 되는데 90+1, 45+2로 되어있어서 int로 형변환하지 못한다. 
# 일단 90+1은 90으로, 45+2는 45로 표현하기로 약속하자.
df3['Minute'] = df3['Minute'].str[0:2]
    
```

```python
# 'Minute 속성의 자료형을 int로 바꾸자
df3['Minute'] = df3['Minute'].astype(int)
```

```python
# 자료형을 확인한다
df3.info()
```

```python
# 0 ~ 3o분은 1, 31분 ~ 60분은 2, 61 ~ 90분은 3으로 판별한다.
# 원래는 15분단위로 판별하려 했으나 그렇게 하면 너무 골 분포도가 세밀어진다. 따라서 분석하는데 어려움이 있다.
# 파생변수 Goal_Division을 만든다.
def func(x):
    if x >= 0 and x <= 30 :
        return 1
    elif x >= 31 and x <= 60:
        return 2
    else:
        return 3
     
    
df3['Goal_Division'] = df3['Minute'].apply(lambda x : func(x))
```

```python
# 'Goal Division' 파생변수가 잘 나오는지 전체 데이터를 확인한다.
df3
```

```python
# 'Goal_Division'과 'Type'를 이용한 이중변수 분석
df4 = df3.groupby('Goal_Division')
```

```python
# 0 ~ 30분에 호날두가 넣은 골의 분포도
df4.get_group(1)['Type'].value_counts()
```

```python
# 31 ~ 60분에 호날두가 넣은 골의 분포도
df4.get_group(2)['Type'].value_counts()
```

```python
# DataFrame를 만들고, heatmap으로 데이터 시각화 한다.

# create Dataframe

df = pd.DataFrame(value,columns=x, index=y)

# heatmap
h = 356
s = 0.78
v = 0.92
 
colors = [
    mcl.hsv_to_rgb((h/360,0,v)),
    mcl.hsv_to_rgb((h/360,0.5,v)),
    mcl.hsv_to_rgb((h/360,1,v))
]

cmap = LinearSegmentedColormap.from_list('my_cmap',colors,gamma=1)

fig, ax = plt.subplots(figsize=(16, 12))    
fig.set_facecolor('white')

ax = sns.heatmap(df,
                 cmap=cmap,            
                 annot=True,
                 annot_kws={"size": 15},             
                 fmt="d",                 
                 linewidths=3)

sns.set(font_scale=1.5)
```
